## OHE - One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python

documents = ["people watch movie", 
             "people watch cricket", 
             "people like movie", 
             "people like cricket"]
```

```python
tokens = [sentence.lower().split() for sentence in documents]
print(tokens)
```

```python
all_words = [[word] for sentence in tokens for word in sentence]
print(all_words)
```

sparse_output=False means the encoder returns a NumPy array instead of a sparse matrix. This makes the output easier

```python
encoder = OneHotEncoder(sparse_output=False)
encoder.fit(all_words)
encoder.categories_
```

```python
print(encoder.categories_[0])
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print(encoded_sentence)
```

## BOW - Bag of Words

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = ["people watch movie and watch movie again", 
             "people watch cricket and watch cricket",
             "people like movie and like movie a lot",
             "people like cricket"]
```

```python
bow = CountVectorizer()
```

```python
bow_matrix =bow .fit_transform(documents)
print("Vocabulary\n",bow.get_feature_names_out())
print("BOW Matrix:\n", bow_matrix.toarray())
```

# Embedding

## Word2Vec
* The model is trained over google news data
* Dimension of model (number of neurons in hidden layer) = 300

```python
from gensim.models import Word2Vec,KeyedVectors
import gensim.downloader as api
```

```python
#skim gram
model = api.load("word2vec-google-news-300")
```

```python
man_embedding = model['man']
print("Dimension of man embedding:", len(man_embedding))
print("man embedding:\n", man_embedding)
```

```python
model.most_similar("man")
```

```python
model.most_similar("google")
```

```python
# Finding similarity between words

print("Similarity between 'king' and 'queen':", model.similarity("king", "queen"))
print("Similarity between 'king' and 'man':", model.similarity("king", "man"))
print("Similarity between 'boy' and 'door':", model.similarity("boy", "door"))
print("Similarity between 'king' and king':", model.similarity("king", "king"))
```

```python
# Finding odd one out in a list of words
model.doesnt_match(["PHP","JAVA","DOG","C++"])
```

```python
# Doing arthmetic operations on word embeddings
vector = model['king'] - model['man'] + model['woman']
model.most_similar([vector], topn=5)
```

```python
# Embedding of a sentence is calculated by averaging the embeddings of all the words in the sentence.

sentence = "people watch movie and watch movie again"
words = sentence.lower().split()
print("Words in the sentence:", words)
word_embeddings = [model[word] for word in words if word in model]
print("dimension of each word embedding:", len(word_embeddings[0]))
```

```python
import numpy as np
sentence_vector =  np.mean(word_embeddings, axis=0)
print("Embedding of the sentence:\n", sentence_vector)
```
