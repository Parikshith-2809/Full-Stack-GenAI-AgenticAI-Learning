```python
from sklearn.preprocessing import OneHotEncoder
import numpy as np
```

```python

documents = ["people watch movie", 
             "people watch cricket", 
             "people like movie", 
             "people like cricket"]
```

```python
tokens = [sentence.lower().split() for sentence in documents]
print(tokens)
```

```python
all_words = [[word] for sentence in tokens for word in sentence]
print(all_words)
```

sparse_output=False means the encoder returns a NumPy array instead of a sparse matrix. This makes the output easier

```python
encoder = OneHotEncoder(sparse_output=False)
encoder.fit(all_words)
encoder.categories_
```

```python
print(encoder.categories_[0])
for sentence in tokens:
    encoded_sentence = encoder.transform([[word] for word in sentence])
    print("\nSentence:", sentence)
    print(encoded_sentence)
```
